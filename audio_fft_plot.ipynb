```python
import os
import matplotlib.pyplot as plt
import librosa, librosa.display
import numpy as np
```

```python
def DoFFT(title):
    # load sounds
    audio, sr = librosa.load(title+".wav")
    plot_magnitude_spectrum(audio, sr, title, 0.1)
```

```python
def plot_magnitude_spectrum(signal, sr, title, f_ratio=1):
    X = np.fft.fft(signal)
    X_mag = np.absolute(X)
    
    plt.figure(figsize=(5, 5))
    
    f = np.linspace(0, sr, len(X_mag))
    f_bins = int(len(X_mag)*f_ratio)  
    
    plt.plot(f[:f_bins],  X_mag[:f_bins])
    plt.xlabel('Frequency (Hz)')
    plt.title(title)
```

```python
DoFFT("Honeypie")
```

```python
DoFFT("Treasure")
```

```python
DoFFT("Paris In The Rain")
```
